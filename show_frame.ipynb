```python
import os, sys
import matplotlib.pyplot as plt
import numpy as np


# load data from csv file
def load_data(file_path):
    data = np.genfromtxt(file_path, delimiter=',', skip_header=1)
    return data


# plot data in 3d
def plot_data(data):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = np.arange(32)
    y = np.arange(24)
    X, Y = np.meshgrid(x, y)
    Z = data
    ax.plot_surface(X, Y, Z, cmap='jet')
    ax.view_init(elev=30, azim=15)
    plt.show()


path = "20200626_154313_mlx90640_01_light_none.csv"
raw = load_data(path)[:, -768:]
```

```python
f0 = raw[85, -768:].reshape(24, 32)
plt.imshow(f0, cmap='jet', interpolation='nearest')
plt.show()

plot_data(f0)

frame = raw[305, -768:].reshape(24, 32)
plt.imshow(frame, cmap='jet', interpolation='nearest')
plt.show()

plot_data(frame)
```

```python
import PIL.Image as Image

img_path = "untitled.png"
img = Image.open(img_path).convert('L')
# reshape image to 24x32
img = img.resize((32, 24))
img = np.array(img)
# normalize image
img = img / 255.0

plt.imshow(img, cmap='jet', interpolation='nearest')
plt.show()

plot_data(img)
```
